```python
#1. How many restaurants for each cuisine type?
fig=px.histogram(zomato_dataset,x='CUSINE TYPE',color='CUSINE TYPE',title='NO.OF RESTAURANTS BY CUISINE TYPE',labels={'CUSINE TYPE':'CUISINE TYPE'})
fig.show()
```

```python
#2.What are the pecentages of restaurants for each rating type?
rating_type_df=zomato_dataset['RATING_TYPE'].value_counts().reset_index()
rating_type_df.rename(columns={'RATING_TYPE':'Rating type','count':'Count of Restaurants'},inplace=True)
```

```python
rating_type_df
```

```python
fig=px.pie(rating_type_df,names='Rating type',values='Count of Restaurants',title='Percentages of Restaurants by Rating type').update_traces(textinfo='percent+label')
fig.show()
```

```python
#3. Which are the top 10 highest rated Seafood restaurants in Mumbai?
seafood_df=zomato_dataset[zomato_dataset['CUSINE_CATEGORY'].str.contains('Seafood')]
seafood_df=seafood_df.sort_values(by='RATING',ascending=False).head(10)
seafood_df
```

```python
#4. Which is the best food truck in Mumbai?
foodtruck_df=zomato_dataset[zomato_dataset['CUSINE TYPE']=='Food Truck']
foodtruck_df=foodtruck_df.sort_values(by='RATING',ascending=False).head(2)
foodtruck_df
```

```python
#5. Which places have the highest rated restaurants for each type of cuisine in Mumbai?
highest_rated_df=zomato_dataset[zomato_dataset['RATING']>=4.5]
highest_rated_df
```

```python
fig=px.histogram(highest_rated_df,x='REGION', color='CUSINE TYPE',title='No. of Best Restaurants in each Region by Cuisine type').update_xaxes(categoryorder='total descending')
fig.show()
```

```python
#6. What is the average price distribution of highest rated restaurant for each cuisine type in Mumbai?
highest_rated_price_df=highest_rated_df.groupby(by=['CUSINE TYPE'])['PRICE'].mean().reset_index()
highest_rated_price_df
```

```python
fig=px.scatter(highest_rated_price_df,x='CUSINE TYPE',y='PRICE',color='CUSINE TYPE',symbol='CUSINE TYPE',
    title='Avg price of highest rated Restaurants by cuisine type').update_traces(marker_size=9)
fig.show()
```

```python
#7. Which areas have a large number of Chinese restaurants?
chinese_rest_df=zomato_dataset[zomato_dataset['CUSINE_CATEGORY'].str.contains('Chinese')]
chinese_rest_df
```

```python
chinese_area_df=chinese_rest_df['REGION'].value_counts().reset_index()
chinese_area_df
```

```python
fig=px.scatter(chinese_area_df,x='REGION',y='count',title='No. of Chinese Restaurants by Region').update_traces(marker_size=7)
fig.show()
```

```python
#8. Is there a relation between price and rating by each Cuisine Type?
price_rating_df=zomato_dataset.groupby(by=['CUSINE TYPE','RATING'])['PRICE'].mean().reset_index()
price_rating_df                                      
```

```python
fig=px.line(price_rating_df,x='RATING',y='PRICE',color='CUSINE TYPE',title='Relation between price and rating for each Cuisine Type')
fig.show()
```

```python
#9. Is there a relation between Region and Price?
region_price_df=zomato_dataset.groupby(by='REGION')['PRICE'].mean().reset_index()
region_price_df
```

```python
fig=px.scatter(region_price_df,x='REGION',y='PRICE',title='Relation between Region and Price')
fig.show()
```

```python
#10. Find the list Affordable restaurants
#Affordable means Low Price and High Rating
max_price=zomato_dataset['PRICE'].max()
aff_price=max_price/4
aff_price
```

```python
aff_price_df=zomato_dataset[zomato_dataset['PRICE']<=1250]
aff_price_df
```

```python
high_rating_df=zomato_dataset[zomato_dataset['RATING']>=4.5]
high_rating_df
```

```python
aff_rest_df=pd.merge(aff_price_df,high_rating_df,how='inner')
aff_rest_df
```

```python
#11. Find the list of most reliable restaurants.
#The most reliable restaurants would have low price, high rate and large no. of votes
mean_votes=zomato_dataset['VOTES'].mean()
mean_votes
```

```python
large_votes_df=zomato_dataset[zomato_dataset['VOTES']>=177]
large_votes_df
```

```python
reliable_rest_df=pd.merge(aff_rest_df,large_votes_df,how='inner')
reliable_rest_df.sort_values(by='PRICE',inplace=True)
reliable_rest_df
```
